```python
import numpy as np
def func(x):
    return x**3+3*x+4
def difdivididas(x, h):
    return (func(x+h)-func(x-h))/(2*h)
def Richardson(x, h):
    return (-func(x+2*h)+8*func(x+h)-8*func(x-h)+func(x-2*h))/(12*h)
print(f"La derivada de la funcion por Diferencias centradas es {difdivididas(2, 0.001)}")
print(f"La derivada de la funcionpor Richardson es: {Richardson(2, 0.001)}")
```

```python
import numpy as np
def func(x):
    return x**3+3*x+4
def difdivididas(x, h):
    f1=func(x+h)
    f2=func(x-h)
    return (f1-f2)/(2*h)


def difdivididasss(x, h, nsimaderivada=1):
    f1=func(x+h)
    f2=func(x-h)
    for i in range(1, nsimaderivada):
        f1=difdivididasss(f1,h,i)
        f2=difdivididasss(f2,h,i)

    return (f1-f2)/(2*h)

print(f"La derivada de la funcion es: {difdivididasss(2, 0.001, 2)}")
```
